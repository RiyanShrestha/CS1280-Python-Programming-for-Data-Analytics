```python
plt.figure(figsize=(10,6))
plt.hist(df['Purchase Amount (USD)'],color='skyblue',bins= 30)
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('frequency')
plt.title('histogram of Purchase Amount (USD)')
plt.show()
```

```python
plt.bar(df['Item Purchased'],df['Review Rating'],color='green')
plt.xlabel('Item Purchased')
plt.ylabel('Review Rating')
plt.title('Bar Chart of Item Purchased vs Age')
plt.show()
```

```python
plt.plot(df['Frequency of Purchases'],df['Review Rating'],marker='o',linestyle='-',color='red')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Review Rating')
plt.title('Scatter Plot of Frequency of Purchases vs Age')
plt.show()
```

```python

plt.scatter(df['Age'], df['Purchase Amount (USD)'], color='blue', marker='o')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.title('Scatter Plot of Age vs Purchase Amount') # Added title for clarity
plt.show()
```

```python
sns.boxplot(x=df['Purchase Amount (USD)'],y=df['Category'],color="pink")
plt.xlabel('Discount Applied')
plt.show()
```

```python
sns.violinplot(x=df['Age'],y=df['Category'],color='yellow')
plt.xlabel('Age')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.show()
```
